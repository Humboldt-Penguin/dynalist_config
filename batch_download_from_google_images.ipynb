Instructions:


1.   Get Google Image data
    * Go on Google Images and search for what you want
    * Right click > "Inspect" (opens a menu on the right side)
    * Select "Network" at the top bar of the menu
    * Select "Img" in the third bar from the top
    * Refresh the Google Search page
    * In the filter box, enter “images?q=”
    * Scroll to the bottom of the Google Search page to get as many results as you'd like
    * In the "Name" column, right click > "Copy" > "Copy all as cURL (bash)"
    * Make a new text file in your computer, paste in the data, and save
2.   Generate PowerShell script by running this code and inputting the text file. 
3. Download resulting `download_images.ps1` from the "Files" menu on the left. Save it in the folder where you'd like to download the images.
4. Shift + right click in the aforementioned folder and select "Open PowerShell window here".
5. Run these two commands: 
    * `Set-ExecutionPolicy -Scope Process -ExecutionPolicy Bypass`
    * `.\download_images.ps1`
6. Enjoy bc this took me 2 hours to figure out.

```python
"from google.colab import files
fin = files.upload()

name = list(fin.keys())[0]
# fin[name].decode("utf-8")
dat0 = fin[name].decode("utf-8").split("\r\n")

dat1 = []
i = 0

for line in dat0:
    if "curl 'https://encrypted" in line:
        dat1.append(line[:line.find("--compressed")-1] + f" -o \"img{i}.jpg\"")
        i += 1

with open('download_images.ps1', 'w') as writefile:
    writefile.write("\n".join(dat1))
```
